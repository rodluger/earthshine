```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plot the labeled map
fig, ax = plt.subplots(1, figsize=(10, 4))

# Plot the weighted solution
img = np.load("../data/hi-contrast-map.npz")["img"]
cmap = plt.get_cmap("Blues_r")
cmap.set_under(cmap(0))
im = ax.imshow(img, cmap=cmap, extent=(-180, 180, -90, 90), origin="lower", vmin=0, vmax=1);
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_yticks([-90, -45, 0, 45, 90]);
ax.set_xticks([-180, -135, -90, -45, 0, 45, 90, 135, 180]);
for tick in ax.xaxis.get_major_ticks() + ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(10)

ax.annotate("Africa", xy=(0.538, 0.47), xycoords="axes fraction", ha="center", fontsize=10, color="k");
ax.annotate("Europa", xy=(0.49, 0.75), rotation=45, xycoords="axes fraction", ha="center", fontsize=8, color="k");
ax.annotate("Amazon", xy=(0.29, 0.485), xycoords="axes fraction", ha="center", fontsize=10, color="k");
ax.annotate("Monsoonia", xy=(0.75, 0.56), xycoords="axes fraction", ha="center", fontsize=10, color="k");
ax.annotate("Aleutia", xy=(0.125, 0.77), rotation=40, xycoords="axes fraction", ha="center", fontsize=10, color="k");
ax.annotate("Austronesia", xy=(0.9, 0.225), xycoords="axes fraction", ha="center", fontsize=8, color="k");

ax.annotate("Indian\nOcean", xy=(0.725, 0.265), xycoords="axes fraction", ha="center", fontsize=10, color="w");
ax.annotate("South\nAtlantic\nOcean", xy=(0.42, 0.34), xycoords="axes fraction", ha="center", fontsize=8, color="w");
ax.annotate("North\nAtlantic\nOcean", xy=(0.34, 0.66), xycoords="axes fraction", ha="center", fontsize=8, color="w");
ax.annotate("Pacific\nOcean", xy=(0.055, 0.39), xycoords="axes fraction", ha="center", fontsize=8, color="w");

# Save
#fig.savefig("../tex/figures/map_labels.pdf", bbox_inches="tight")
```
